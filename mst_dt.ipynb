```python
import pandas as pd
pd.options.mode.chained_assignment = None

import os
```

```python
df_list = []
files = os.listdir(os.getcwd())
files_csv = [f for f in files if f[:5] == 'olist']

for i in range(len(files_csv)):
    df_list.append(pd.read_csv(files_csv[i]))
```

```python
for i in range(len(files_csv)):
    df_list[i] = df_list[i].iloc[:1000]
    df_list[i].to_csv(str(files_csv[i]))
```
